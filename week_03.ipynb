# CH4. 선택하기: if

## 중첩 조건문
- nested conditinoal
- if 블록 안에 또 다른 if 블록이 있는 경우
- 들여쓰기로 논리를 구분한다.

```python
# 0 < x < 10
x = 3
if x > 0:
    if x < 10:
        print("Block A")
    else:
        print("Block B")
else:
    print("Block C")
```

## 논리연산자
- 비교연산자가 여러번 사용될 때 사용하는 것
- and, or, not
    - A and B: A도 참이고, B도 참일 때 True
        - 예: A & B
    - A or B: A 혹은 B, 둘 중 하나만 참이면 True
        - 예: A | B
    - not A: A가 False -> True, True -> False
- 비교연산자의 괄호는 가독성을 위해 (optional)
- 파이썬에서만 가능한 표현: 0 < x < 10

```python
if (x > 0) and (x < 10): #논리 연산자
    print("Block A")
else:
    print("Block B")
```

## in
- membership operator
- 어떤 요소가 그 안에 있는지 확인하고 싶다
- 비슷한 조건이 여러번 반복될 때 사용 가능

```python
#리스트 요소에 letter와 일치하는 것이 있는가
letter = 'k'
if letter in ['a', 'e', 'i', 'o', 'u']:
    print("Yes")
else:
    print("No")
    
# word에 모음이 있는지 확인(X)
word = 'abcde'
if word in ['a', 'e', 'i', 'o', 'u']:
    print("Yes")
else:
    print("No")
```

## 바다코끼리 연산자
- walrus operator
- 할당과 테스트를 동시에 하는 기능
- :=

```python
tweet_limit = 200
tweet_string = 'blah' * 49 #공백없이 스트링 4*49

#if문
diff = tweet_limit - len(tweet_string)
if diff > 0:
    print("가능")
else:
    print("불가능")
    
#바다코끼리
if diff := tweet_limit - len(tweet_string) > 0:
        print("가능")
```

## Quiz
- 윤년을 구하는 코드를 짜보자
    - 연도가 4로 나눠떨어지면 윤년
    - 100으로 나눠떨어지면 안된다
    - 400으로 나뉘어떨어지면 된다
    
- year를 입력받는다.
- 출력='{}년은 윤년입니다' / "{}년은 윤년이 아닙니다."

```python
year = int(input('연도를 입력하세요:'))
if year % 4 == 0 and year % 100 != 0 or year % 400 == 0:
    print(f'{year}년은 윤년입니다.')
else:
    print(f'{year}년은 윤년이 아닙니다.')
```

## Slicing
- 지정 범위의 문자열을 추출하는 것
- 문법: 대괄호 + 오프셋 이용[offset]
    - [start:end:step]
- 오프셋은 0부터 시작
- start 포함, end 불포함 (크거나 같고, 작다)
- 0일 경우 생략한다.
    - 예: ~4까지 출력하라 => [:5]
    - 처음부터 끝까지 2스텝씩 => [::2]
    - 처음부터 끝까지 역순으로 찍기 => [::-1]
- 오프셋은 거꾸로도 가능하다
0 1 2 ..
h e l l o
-5     -1

```python
sent = 'hello world I\'m a boy'
print(sent[1]) #2번째 글자
print(sent[3:4]) #3만 나옴
print(sent[:4]) # 처음부터 4까지
print(sent[::2]) #처음부터 끝까지 2스텝씩 출력하기(공백까지 포함한다)
print(sent[::-1]) #반대로
# hello[2] = 'L' #error
```
